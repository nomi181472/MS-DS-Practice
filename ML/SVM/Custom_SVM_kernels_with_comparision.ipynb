##2 Kernel Matrix Generation & 3 Implementation of SVM using Custom Kernel Matrix:

```python
from sklearn.metrics import roc_curve

from sklearn.svm import SVC
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.inspection import DecisionBoundaryDisplay
class AnalyzeResult:

  def __init__(self,y_test,y_pred):
    self.y_test=y_test
    self.y_pred=y_pred

  def analyze(self,):
    print('Model accuracy score with custom kernel  {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
    cm = confusion_matrix(self.y_test, self.y_pred)


# Assuming you have your confusion matrix as a numpy array named 'cm'
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    #print('Confusion matrix\n\n', cm)
    #print('\nTrue Positives(TP) = ', cm[0,0])

    #print('\nTrue Negatives(TN) = ', cm[1,1])

    #print('\nFalse Positives(FP) = ', cm[0,1])

    #print('\nFalse Negatives(FN) = ', cm[1,0])
    TP = cm[0,0]
    TN = cm[1,1]
    FP = cm[0,1]
    FN = cm[1,0]
    print()
    print(classification_report(self.y_test, self.y_pred))
    classification_error = (FP + FN) / float(TP + TN + FP + FN)

   # print('Classification error : {0:0.4f}'.format(classification_error))
    recall = TP / float(TP + FN)

    #print('Recall or Sensitivity : {0:0.4f}'.format(recall))
    true_positive_rate = TP / float(TP + FN)


    #print('True Positive Rate : {0:0.4f}'.format(true_positive_rate))
    false_positive_rate = FP / float(FP + TN)


    #print('False Positive Rate : {0:0.4f}'.format(false_positive_rate))
    specificity = TN / (TN + FP)

    #print('Specificity : {0:0.4f}'.format(specificity))
    metrics = ["Classification Error", "Recall","true_positive_rate","False Positive Rate","Specificity"]
    values = [classification_error,recall, true_positive_rate, false_positive_rate,specificity]

    # Create a bar graph
    plt.figure(figsize=(8, 6))
    plt.bar(metrics, values, color=['coral', 'lightskyblue', 'lightgreen'])
    plt.xlabel("Metrics")
    plt.ylabel("Value")
    plt.title("Evaluation Metrics")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()

    # Display the graph
    plt.show()
  def show_roc_curve(self):

    fpr, tpr, thresholds = roc_curve(self.y_test, self.y_pred)

    plt.figure(figsize=(6,4))

    plt.plot(fpr, tpr, linewidth=2)

    plt.plot([0,1], [0,1], 'k--' )

    plt.rcParams['font.size'] = 12

    plt.title('Prediction')

    plt.xlabel('False Positive Rate (1 - Specificity)')

    plt.ylabel('True Positive Rate (Sensitivity)')

    plt.show()
  @staticmethod
  def plot_decision_boundry(clf,X,Y,kernel_name):

    ax = plt.gca()
    DecisionBoundaryDisplay.from_estimator(
        clf,
        X,
        cmap=plt.cm.Paired,
        ax=ax,
        response_method="predict",
        plot_method="pcolormesh",
        shading="auto",
    )

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Paired, edgecolors="k")
    plt.title(f"Classification with kernel: {kernel_name}")
    plt.axis("tight")
    plt.show()
```

```python
from sklearn.metrics.pairwise import laplacian_kernel,polynomial_kernel,rbf_kernel,sigmoid_kernel,cosine_similarity

def MyCustomKernel(X, Y,gamma=0.5):
    laplacian = laplacian_kernel(X, Y,gamma=gamma)
    cosine=cosine_similarity(X, Y)
    return laplacian_kernel(X,Y,gamma=gamma)*laplacian_kernel(X,Y,gamma=gamma)*laplacian_kernel(X,Y,gamma=gamma)


svm = SVC(kernel=MyCustomKernel)
svm.fit(X_train,y_train)

y_pred=svm.predict(X_test)
AnalyzeResult(y_test, y_pred).analyze()
AnalyzeResult(y_test, y_pred).show_roc_curve()
```

##4 Implementation of SVM using Existing Kernel Matrices & 5. Comparison and Analysis::

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import time
stats_df=pd.DataFrame(columns=["kernels","accuracy","computational_time"])
all_kernels_clf={}

for kernel in [MyCustomKernel,"linear","rbf","poly","linear","sigmoid"]:
  t1=time.time()
  dic={}
  model=SVC(kernel=kernel)

  score=cross_val_score(model,X,y,cv=5,scoring='accuracy').mean()
  if type(kernel)==str:
    dic["kernels"]=kernel
    all_kernels_clf[kernel]=model
  else:
    dic["kernels"]="customkernel"
    all_kernels_clf["customkernel"]=model
  dic["computational_time"]=time.time()-t1
  dic["accuracy"]=score
  stats_df = stats_df.append(dic, ignore_index=True)

print(stats_df.head())
```

```python

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.bar(stats_df.kernels, stats_df.computational_time, color='orange', alpha=0.7)
ax1.set_xlabel('Kernel Names')
ax1.set_ylabel('Computational Time (sec)')
ax1.set_title('Computational Time')
ax1.grid(axis='y', linestyle='--', alpha=0.7)
ax2.bar(stats_df.kernels, stats_df.accuracy, color='yellow', alpha=0.7)
ax2.set_xlabel('Kernel Names')
ax2.set_ylabel('Accuracy (%)')
ax2.set_title('Accuracy')
ax2.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

##6 Optimization and Fine-tuning:

```python
from sklearn.model_selection import GridSearchCV
params = {
    'kernel': [MyCustomKernel, 'linear', 'rbf', 'poly', 'sigmoid'],
    'degree': [2,3],
    'gamma': [ 0.01, 0.5],
    'C': [ 0.5,0.1, ]
}

svm = SVC()
grid_search = GridSearchCV(estimator=svm, param_grid=params, cv=10)
grid_search.fit(X_train,y_train)
best_params = grid_search.best_params_
print("Best Parameters:", best_params)
best_estimator = grid_search.best_estimator_
y_pred = best_estimator.predict(X_test)
print("After fine tuning.")
AnalyzeResult(y_test, y_pred).analyze()
AnalyzeResult(y_test, y_pred).show_roc_curve()



```

```python

from sklearn.model_selection import train_test_split, cross_val_predict
from sklearn.svm import SVC
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier

y = label_binarize(y, classes=[0, 1, 2])
n_classes = y.shape[1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
kernel_types = [MyCustomKernel,"linear","rbf","poly","linear","sigmoid"]

plt.figure()
lw = 2
for kernel in kernel_types:
    classifier = OneVsRestClassifier(SVC(kernel=kernel, probability=True, random_state=0))
    y_score = cross_val_predict(classifier, X_train, y_train, cv=10, method='accuracy')
    fpr, tpr, _ = roc_curve(y_train.ravel(), y_score.ravel())
    roc_auc = auc(fpr, tpr)
    kernel_type=kernel
    if type(kernel)!=str:
      kernel_type="custom_kernel"
    plt.plot(fpr, tpr, lw=lw, label='ROC curve (area = %0.2f) - Kernel: %s' % (roc_auc, kernel_type))

plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.legend(loc="lower right")
plt.show()
```
