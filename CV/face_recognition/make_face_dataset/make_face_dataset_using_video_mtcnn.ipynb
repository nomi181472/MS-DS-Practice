```python
import cv2
import os
from facenet_pytorch import MTCNN,InceptionResnetV1

def capture_images(cp, name,video_path):
    cam = cv2.VideoCapture(video_path)
    cv2.namedWindow("Capture Face")

    face_id = name.split('/')[-1]  # Get the last part of the path for the face_id
    count = 0
    detector = MTCNN(image_size=(128,128),keep_all=True)  # Initialize MTCNN detector
    
    while True:
        ret, frame = cam.read()
        if not ret:
            print("Failed to grab frame.")
            break

        # Detect faces using MTCNN
        results,_ = detector.detect(frame)
        if results is not None:
            for result in results:
                
                x, y, w, h = map(int, result)
                cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)
                count += 1
                save_p = os.path.join(cp, f"{name}_{count}_new_images1.jpg")  # Append .jpg here
                face = frame[y:y + h, x:x + w]
                face=cv2.resize(face,(128,128))
                print(f"face_shape:{face.shape}")
                cv2.imwrite(save_p, face)  # Save the image

        cv2.imshow("Capture Face", frame)
        k = cv2.waitKey(1)
        if k % 256 == 27:  # ESC pressed
            break

    cam.release()
    cv2.destroyAllWindows()

# Call the function
data = "data_set_not_for_upload"
if not os.path.exists(data):
    os.makedirs(data)

employee_name = "4_basit"
path = os.path.join(data, employee_name)
if not os.path.exists(path):
    os.makedirs(path)
video_path="../videos/Abdul Basit.mp4"
print(path)
capture_images(path, employee_name,video_path)
```

```python
import cv2
import torch
from facenet_pytorch import MTCNN
from PIL import Image, ImageOps
import numpy as np
import os

def process_video(video_path, save_path, person_name, prob_threshold=0.98):
    # Create a video capture object
    cap = cv2.VideoCapture(video_path)
    
    # Initialize MTCNN for face detection
    mtcnn = MTCNN(keep_all=True, device='cpu')  # You can switch to 'cuda' if you have a GPU
    
    # Make sure save_path exists
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    
    frame_count = 0
    saved_images = 0
    
    while cap.isOpened():
        ret, frame = cap.read()
        
        if not ret:
            break
        
        frame_count += 1
        # Convert the frame to RGB
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Detect faces and probabilities in the frame
        boxes, probs = mtcnn.detect(rgb_frame)
        
        # If faces are found, draw rectangles and save the images with high probability
        if boxes is not None and probs is not None:
            for i, (box, prob) in enumerate(zip(boxes, probs)):
                if prob >= prob_threshold:
                    # Draw rectangle around the face
                    cv2.rectangle(frame, 
                                  (int(box[0]), int(box[1])), 
                                  (int(box[2]), int(box[3])), 
                                  (0, 255, 0), 2)
                    
                    # Extract the face from the frame
                    face = rgb_frame[int(box[1]):int(box[3]), int(box[0]):int(box[2])]
                    
                    # Convert face to PIL image
                    face_pil = Image.fromarray(face)
                    
                    # Get the aspect ratio and add padding
                    face_pil = add_padding_to_square(face_pil, target_size=128)
                    
                    # Save the face image with person's name
                    face_save_path = os.path.join(save_path, f'z_{person_name}_{frame_count}_{i}_prob_{str(prob)}.png')
                    face_pil.save(face_save_path)
                    print(face_save_path)
                    saved_images += 1
        
        # Display the frame with rectangles
        #cv2.imshow('Video with Face Detection', frame)
        
        # Exit on 'q' key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    
    # Release the capture and destroy any OpenCV windows
    cap.release()
    cv2.destroyAllWindows()
    
    print(f"Total frames processed: {frame_count}")
    print(f"Total face images saved: {saved_images}")

def add_padding_to_square(image, target_size=128):
    # Calculate padding to maintain aspect ratio
    width, height = image.size
    max_dim = max(width, height)
    
    # Create new image with square dimensions and paste original
    padded_image = ImageOps.pad(image, (max_dim, max_dim), color=(0, 0, 0))
    
    # Resize to target size
    padded_image = padded_image.resize((target_size, target_size))
    
    return padded_image

# Example usage:
# process_video('path_to_video.mp4', 'path_to_save_images', 'John_Doe')
```
